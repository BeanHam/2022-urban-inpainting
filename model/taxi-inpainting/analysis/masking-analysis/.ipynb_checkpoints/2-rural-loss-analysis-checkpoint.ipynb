```python
import numpy as np
import pandas as pd
import torch
import torch.nn as nn
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
result = pd.read_csv(f'rural_target_results.csv')
biased = result.iloc[::2, :]
random = result.iloc[1::2, :]

biased = biased.hole_l1_output.values
random = random.hole_l1_output.values
T = [1,2,3,5,7,10,15]
```

```python
color_palette = sns.color_palette('Paired')
dot_size = 80
x_tick_size = 16
y_tick_size = 16
legend_size = 18
label_size = 20
line_with = 4

## first plot
plt.figure(figsize=(12,5))
plt.scatter(T, biased, color='black',s=dot_size)
plt.plot(T, biased, label='Trained With Biased Masking', color='black', linewidth=line_with, linestyle='solid')

plt.scatter(T, random,color='grey',s=dot_size)
plt.plot(T, random, label = 'Trained With Random Masking', color='grey', linewidth=line_with, linestyle='dashed')

plt.legend(fontsize=legend_size, loc=1)
plt.grid(True)
#plt.ylim()
plt.ylabel('L1 Error (Hole)\n -- Sparse Region', size=label_size)
plt.xlabel('Temporal Dimension', size=label_size)
plt.xticks([1,2,3,5,7,10,15], ['T=1','T=2','T=3','T=5','T=7', 'T=10', 'T=15'],size=x_tick_size)
plt.yticks(size=y_tick_size)
pass
```
