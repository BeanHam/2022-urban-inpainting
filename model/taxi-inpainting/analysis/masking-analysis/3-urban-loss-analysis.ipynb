```python
import numpy as np
import pandas as pd
import torch
import torch.nn as nn
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
result = pd.read_csv(f'urban_target_results.csv')
biased = result.iloc[::4, :]
biased_mape = result.iloc[1::4, :]
random = result.iloc[2::4, :]
random_mape = result.iloc[3::4, :]

biased = biased.hole_l1_output.values
biased_mape = biased_mape.hole_l1_output.values
random = random.hole_l1_output.values
random_mape = random_mape.hole_l1_output.values
T = [1,2,3,5,7,10,15]
```

```python
color_palette = sns.color_palette('Paired')
dot_size = 80
x_tick_size = 16
y_tick_size = 16
legend_size = 18
label_size = 20
line_with = 4

## first plot
plt.figure(figsize=(12,5))
plt.scatter(T, biased, color=color_palette.__getitem__(0),s=dot_size)
plt.plot(T, biased, label='Biased Masking', color=color_palette.__getitem__(0), linewidth=line_with)

plt.scatter(T, biased_mape, color=color_palette.__getitem__(1),s=dot_size)
plt.plot(T, biased_mape, label='Biased Masking, MAPE', color=color_palette.__getitem__(1), linewidth=line_with)

plt.scatter(T, random,color=color_palette.__getitem__(4),s=dot_size)
plt.plot(T, random, label = 'Random Masking', color=color_palette.__getitem__(4), linewidth=line_with)

plt.scatter(T, random_mape,color=color_palette.__getitem__(5),s=dot_size)
plt.plot(T, random_mape, label = 'Random Masking, MAPE', color=color_palette.__getitem__(5), linewidth=line_with)

plt.legend(fontsize=legend_size, loc=1)
plt.grid(True)
plt.ylim(16)
plt.ylabel('L1 Error (Hole)\n -- Sparse Region', size=label_size)
plt.xlabel('Temporal Dimension', size=label_size)
plt.xticks([1,2,3,5,7,10,15], ['T=1','T=2','T=3','T=5','T=7', 'T=10', 'T=15'],size=x_tick_size)
plt.yticks(size=y_tick_size)
pass
```
