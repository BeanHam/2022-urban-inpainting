```python
palettedata = [
    0,0,0, ## black
    255,235,205, ## mistyrose
    255,160,122, ## blanchedAlmond
    255,0,0, ## orangered
]

root="D:/nyc_taxi/train"
background = Image.open("C:/Users/binha/Downloads/nyc.PNG").resize((64,64)).convert("RGBA")
pred = Image.open(root+f'/13.png')
pred.putpalette(palettedata*64)
pred = pred.convert("RGBA")

new_img = Image.blend(pred, background, 0.25 )
plt.figure(figsize=(12,8))
plt.imshow(new_img)
pass
```
