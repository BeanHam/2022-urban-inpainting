```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def loss(x0, x1, beta, lamda, offset=1e-3):
    delta_x = x1-x0
    l1 = delta_x
    l2 = ((delta_x + offset) /(np.min((x0,x1),axis=0)+offset))**lamda
    #l1 = l1/sum(l1)
    #l2 = l2/sum(l2)
    l1 = l1*beta
    l2 = l2*(1-beta)
    return l1, l2, l1+l2
```

$$\ell = \Sigma_{i=1}^{n} \beta * ||y_i-x_i||_1 + (1-\beta)* \bigg[\frac{x_i - y_i}{min(x_i, y_i)}\bigg]^\lambda$$

$$0 \le \beta \le 1$$

$$\lambda \ge 1$$
